```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets
from IPython.display import display
import io

# Віджет для завантаження файлу
uploader = widgets.FileUpload(
    accept='.csv',
    multiple=False,
    description='Завантажити CSV'
)

# Функція для обробки завантаженого файлу
def explore_dataset(change):
    try:
        if uploader.value:
            uploaded_file = list(uploader.value.values())[0]
            global df
            df = pd.read_csv(io.BytesIO(uploaded_file['content']))

            # 1. Перегляд перших рядків
            print("\nПерші 5 рядків датасету:")
            display(df.head())

            # 2. Інформація про стовпці та типи
            print("\nІнформація про датасет:")
            display(df.info())

            # 3. Пропущені значення
            print("\nПропущені значення:")
            missing = df.isnull().sum()
            display(missing)
            print("\nВідтік серед пропущених значень у reamining_contract:")
            display(df[df['reamining_contract'].isnull()]['churn'].mean())

            # 4. Базові статистики для числових стовпців
            print("\nОпис числових стовпців:")
            display(df.describe())

            # 5. Баланс класів у churn
            print("\nБаланс класів у churn:")
            display(df['churn'].value_counts(normalize=True))
            plt.figure(figsize=(6, 4))
            sns.countplot(x='churn', data=df)
            plt.title('Баланс класів у churn')
            plt.show()

            # 6. Розподіл числових змінних
            numeric_cols = ['subscription_age', 'bill_avg', 'download_avg', 'upload_avg']
            for col in numeric_cols:
                plt.figure(figsize=(8, 5))
                sns.histplot(data=df, x=col, hue='churn', kde=True)
                plt.title(f'Розподіл {col} за churn')
                plt.show()

            # 7. Кореляційна матриця
            print("\nКореляційна матриця:")
            plt.figure(figsize=(10, 8))
            sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
            plt.title('Кореляційна матриця')
            plt.show()

            # 8. Виявлення викидів через IQR
            for col in numeric_cols:
                Q1 = df[col].quantile(0.25)
                Q3 = df[col].quantile(0.75)
                IQR = Q3 - Q1
                outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
                print(f"\nКількість викидів у {col}: {len(outliers)}")
                print(f"Частка churn=1 серед викидів у {col}: {outliers['churn'].mean():.2f}")
                plt.figure(figsize=(8, 5))
                sns.boxplot(x='churn', y=col, data=df)
                plt.title(f'Boxplot для {col} за churn')
                plt.show()

            # 9. Перевірка від’ємних subscription_age
            negative_age = df[df['subscription_age'] < 0]
            print("\nКількість рядків із від’ємним subscription_age:", len(negative_age))
            print("Приклади:", negative_age[['subscription_age', 'churn']].head())

            # 10. Групи за reamining_contract
            df['contract_group'] = pd.cut(df['reamining_contract'],
                                          bins=[-float('inf'), 0, 1, float('inf')],
                                          labels=['Без контракту', 'До 1 року', 'Понад 1 рік'])
            print("\nВідтік за групами контракту:")
            contract_churn = df.groupby('contract_group')['churn'].mean()
            display(contract_churn)
            plt.figure(figsize=(8, 5))
            sns.barplot(x='churn', y='contract_group', data=df)
            plt.title('Відтік за групами контракту')
            plt.show()

            # 11. Комбінація TV і movie підписки
            df['tv_and_movie'] = (df['is_tv_subscriber'] & df['is_movie_package_subscriber']).astype(int)
            print("\nВідтік за комбінацією TV і movie:")
            tv_movie_churn = df.groupby('tv_and_movie')['churn'].mean()
            display(tv_movie_churn)
            plt.figure(figsize=(6, 4))
            sns.barplot(x='churn', y='tv_and_movie', data=df)
            plt.title('Відтік за комбінацією TV і movie')
            plt.show()

            # 12. Вплив service_failure_count
            print("\nВідтік за кількістю збоїв:")
            service_failure_churn = df.groupby(pd.cut(df['service_failure_count'], bins=[-1, 0, 1, float('inf')]))['churn'].mean()
            display(service_failure_churn)
            plt.figure(figsize=(8, 5))
            sns.barplot(x='churn', y=pd.cut(df['service_failure_count'], bins=[-1, 0, 1, float('inf')]), data=df)
            plt.title('Відтік за кількістю збоїв')
            plt.show()

        else:
            print("Файл не завантажено!")
    except Exception as e:
        print(f"Помилка: {e}")

# Прив'язуємо функцію до віджета
uploader.observe(explore_dataset, names='value')
display(uploader)
```
